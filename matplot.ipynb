```python
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('sine wave')
plt.show()
```
